```python

#question 1
import numpy as np
x=np.random.rand(10,1)
print("Formed Matrix Is \n",x)

print("Mean Of The Above Matrix Is =",np.mean(x))
```

```python

#Question 2

x=np.random.rand(20,1)
print("Formed Matrix Is \n",x)
print("Variance of the formed matrix is =",np.var(x))
print("Standard deviation of the formed matrix is =",np.std(x))
```

```python

#Question 3

A=np.random.rand(10,20)
B=np.random.rand(20,25)

x=np.matmul(A,B)
print("Matrix Formed After Multiplication\n",x)

print("Sum Of Newly Formed Matrix = ",np.sum(x))
```
